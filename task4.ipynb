```python
#Lorenz system from worksheet 3

t_start, t_end = 0, 100
t = np.linspace(t_start, t_end, 5000)

def lorenz(t, x0):
    """
    t: independent variable (time)
    state: initial values
    
    returns values for x, y and z according to the lorenz system
    """
    x, y, z = x0
    return sigma * (y - x), x * (rho - z) - y, x * y - beta * z
```

```python
#setting up parameters
rho = 28.0
sigma = 10.0
beta = 8.0 / 3.0
#initialize x_0 vector
x0 = [10, 10, 10]    
#solve lorenz attractor using solve_ivp
sol1 = solve_ivp(lorenz, [t_start, t_end], x0, t_eval=t)
# Plot system
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(sol1.y[0], sol1.y[1], sol1.y[2], 'r-', linewidth=0.1, label='x0 = '+str(x0))
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title(r'$\rho=$'+str(rho) + r' $\sigma=$'+str(sigma) + r' $\beta=$'+str(round(beta,2)))
ax.grid()
plt.draw()
```

```python
# Lag x variable
delta_t = 2
lag1 = delta_t
lag2 = 2*delta_t
x1 = sol1.y[0]

length = x1.shape[0]
x2 = x1[lag1:]
x3 = x1[lag2:]
#Plot system
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(x1[:length-lag2], x2[:length-lag2], x3, 'r-', linewidth=0.5, label='x0 = '+str(x0))
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("x3")
ax.set_title(r'$\rho=$'+str(rho) + r' $\sigma=$'+str(sigma) + r' $\beta=$'+str(round(beta,2)))
ax.grid()
plt.draw()
```

```python
#Same for y variable
delta_t = 2
lag1 = delta_t
lag2 = 2*delta_t
y1 = sol1.y[1]

length = y1.shape[0]
y2 = y1[lag1:]
y3 = y1[lag2:]

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(y1[:length-lag2], y2[:length-lag2], y3, 'r-', linewidth=0.5, label='x0 = '+str(x0))
ax.set_xlabel("y1")
ax.set_ylabel("y2")
ax.set_zlabel("y3")
ax.set_title(r'$\rho=$'+str(rho) + r' $\sigma=$'+str(sigma) + r' $\beta=$'+str(round(beta,2)))
ax.grid()
plt.draw()
```
