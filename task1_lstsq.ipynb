```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


data_raw_linear = np.loadtxt("linear_function_data.txt", delimiter=' ')
data_raw_nonlinear = np.loadtxt('nonlinear_function_data.txt', delimiter=' ')
```

```python
def approx_linear(data_in):
    """
    Input: numpy array (x,y) as (rows, columns)
    Calculates m,c (A,C) with least squares
    """
    x = data_in[:,0]
    y = data_in[:,1]
    A = np.vstack([x, np.ones(len(x))]).T
    m, c = np.linalg.lstsq(A, y, rcond=None)[0]
    
    plt.plot(x, y, 'o', label='Original data', markersize=10)
    plt.plot(x, m*x + c, 'r', label='Fitted line')
    plt.legend()
    plt.show()
    
    return
```
