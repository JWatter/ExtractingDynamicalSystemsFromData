```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from utils import *
from mpl_toolkits.mplot3d import Axes3D

windows = 350
```

```python
utilization_path = "./data/MI_timesteps.txt"
utilization_data = pd.read_csv(utilization_path, header='infer', delimiter=" ", skiprows=range(1, 1001))
```

## First Part

```python
col = []
for i in range(1, 10):
    col.append(np.array(utilization_data[utilization_data.columns[i]]))
```

```python
# plot original dataset
plt.plot(utilization_data['timeStep'], col[0])
plt.xlabel('t')
plt.ylabel('PID5')
plt.show
```

```python
# create windows of the first three measurementareas
def rolling_window(a, window):
    shape = a.shape[:-1] + (a.shape[-1] - window, window + 1)
    strides = a.strides + (a.strides[-1],)
    return np.lib.stride_tricks.as_strided(a, shape=shape, strides=strides)

PID5_slided = rolling_window(col[0], windows)
PID6_slided = rolling_window(col[1], windows)
PID7_slided = rolling_window(col[2], windows)
```

```python
data =  np.concatenate((PID5_slided, PID6_slided, PID7_slided),axis=1)
data_norm = (data - data.mean(axis=0)) / data.std(axis=0)
```

```python
# apply PCA with 3 components
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
projected = pca.fit_transform(data_norm)
pca.set_params(copy=True)

explained_variance_sk = pca.explained_variance_ratio_
expl_var_cumsum_sk = np.cumsum(explained_variance_sk)
```

```python
# plot embedding space
fig = plt.figure()
ax=fig.add_subplot(111,projection="3d")
ax.scatter(projected[:, 0], projected[:, 1], projected[:, 2])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('Plot of PCA embedding space')
fig.tight_layout()
```

## Part 2

```python
for i in range(9):
    fig = plt.figure()
    ax=fig.add_subplot(111,projection="3d")
    ax.scatter(projected[:, 0], projected[:, 1], projected[:, 2], s=1,c=rolling_window(col[i], windows)[:, 0])
    ax.set_xlabel('PC1')
    ax.set_ylabel('PC2')
    ax.set_zlabel('PC3')
    ax.set_title('PCA embedding space colored by column' + str(i+2))
    fig.tight_layout()
```

```python
# Approximate vectoe feild using the whole projected dataset

x0 = projected[:-1,:]
x1 = projected[1:,:]
del_t = 1 # set the delta t to 1 timestep
vec = (x1 - x0) / del_t

Approximator = NonlinearApprox(400,12)
v_hat = Approximator.radial_approx(x0, vec)
x1_hat = v_hat * del_t + x0
mse = np.square(x1 - x1_hat).mean()
print("Mean Squared Error: ", mse)
```

```python
# Approximate vectoe feild using a single loop of projected data
x0_sing = projected[:1993,:]
x1_sing = projected[1:1994,:]
del_t = 1 # set the delta t to 1 timestep
vec_sing = (x1_sing - x0_sing) / del_t

Approximator_sing = NonlinearApprox(400,12)
v_hat = Approximator_sing.radial_approx(x0_sing, vec_sing)
x1_sing_hat = v_hat * del_t + x0_sing
mse = np.square(x1_sing - x1_sing_hat).mean()
print("Mean Squared Error: ", mse)
```

## Part 4

```python
# for 7 days contain 15001 timestep, 14 days should loop over 30002 timesteps 
T = 30002
```

```python
# create prediction over 14 days using the whole projected dataset (this take ages to run)
x_hat_pred = [x0] 
for _ in np.arange(T, step=del_t):
    v_hat = Approximator.predict(x_hat_pred)
    x_hat = v_hat * del_t + x_hat_pred
    x_hat_pred = x_hat
```

```python
# plot embedding space after looping 14 days
fig = plt.figure()
ax=fig.add_subplot(111,projection="3d")
ax.scatter(x_hat_pred[:, :, 0], x_hat_pred[:, :, 1], x_hat_pred[:, :, 2], s=1, c=rolling_window(col[0], windows)[:13650, 0])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('Predicted embedding space after 14 days')
fig.tight_layout()
```

```python
# create prediction over 14 days using a single loop of projected
x_hat_pred = [x0_sing] 
for _ in np.arange(T, step=del_t):
    v_hat = Approximator_sing.predict(x_hat_pred)
    x_hat = v_hat * del_t + x_hat_pred
    x_hat_pred = x_hat
    
# plot embedding space after looping 14 days
fig = plt.figure()
ax=fig.add_subplot(111,projection="3d")
ax.scatter(x_hat_pred[:, :, 0], x_hat_pred[:, :, 1], x_hat_pred[:, :, 2], s=1, c=rolling_window(col[0], windows)[:1993, 0])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('Predicted embedding space after 14 days')
fig.tight_layout()
```
