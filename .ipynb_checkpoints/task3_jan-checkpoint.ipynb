```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from utils import *
```

```python
# prepare data
x0_f = "./data/nonlinear_vectorfield_data_x0.txt"
x1_f = "./data/nonlinear_vectorfield_data_x1.txt"
x0 = pd.read_csv(x0_f, header=None, delimiter=" ").values
x1 = pd.read_csv(x1_f, header=None, delimiter=" ").values
```

```python
# Part1: assume small delta t
del_t = 0.1
vec = (x1 - x0) / del_t
```

```python
Approximator = LinearApprox()
v_hat = Approximator.linear_approx(x0, vec)
```

```python
x1_hat = v_hat * del_t + x0
mse = np.square(x1 - x1_hat).mean()
# Remarkable that the MSE is almost invariant to the time step delta t.
# This is due to that fact that delta t works as a sacaling only.
print("Mean Squared Error: ", mse)
```

```python
# Part 2
Approximator = NonlinearApprox(1000,1.5)
delt_t = 0.1
```

```python
v_hat = Approximator.radial_approx(x0, vec)
x1_hat = v_hat * delt_t + x0
mse = np.square(x1 - x1_hat).mean()
# Remarkable that the MSE is almost invariant to the time step delta t.
# This is due to that fact that delta t works as a sacaling only.
print("Mean Squared Error: ", mse)
```

```python
# Weird streamplot
y = np.linspace(-4.5, 4.5, 40)
x = np.linspace(-4.5, 4.5, 50)
matrix = np.array(np.meshgrid(x, y)).T.reshape(-1, 2)
predict = Approximator.predict(matrix)
dx = predict[:, 0].reshape(y.shape[0], x.shape[0])
dy = predict[:, 1].reshape(y.shape[0], x.shape[0])

plt.figure()
a1 = plt.subplot()
a1.streamplot(x, y, dx, dy)
plt.show()
```
