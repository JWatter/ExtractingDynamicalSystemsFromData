```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from utils import *
```

## First Part

```python
linear_path = "./data/linear_function_data.txt"

linear_data = pd.read_csv(linear_path, header=None, delimiter=" ").values
x = linear_data[:, 0]
y = linear_data[:, 1]
A = np.vstack([x, np.ones(len(x))]).T

linear_approx = LinearApprox()
linear_approx.fit(A, y)
predict = linear_approx.predict(linear_data)

a, b = linear_approx.get_coe()
print('coefficients:', a)
print('bias:', b)

fig = plt.figure()
plt.plot(x, y, 'o', label='Original Data')
plt.plot(x, predict, label="Linear Approximation")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Approximate with a Linear Function')
plt.legend()
plt.show()
```

## Second Part

```python
nonlinear_path = "./data/nonlinear_function_data.txt"

nonlinear_data = pd.read_csv(nonlinear_path, header=None, delimiter=" ").values
x = nonlinear_data[:, 0]
y = nonlinear_data[:, 1]
A = np.vstack([x, np.ones(len(x))]).T

linear_approx = LinearApprox()
linear_approx.fit(A, y)
predict = linear_approx.predict(nonlinear_data)

a, b = linear_approx.get_coe()
print('coefficients:', a)
print('bias:', b)

fig = plt.figure()
plt.plot(x, y, 'o', label='Original Data')
plt.plot(x, predict, '.', label="Linear Approximation")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Approximate with a Nonlinear Function')
plt.legend()
plt.show()
```
