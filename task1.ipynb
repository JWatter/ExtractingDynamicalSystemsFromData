```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from utils import *
```

## First Part

```python
# Plot linear approximation against original data (linear)
linear_path = "./data/linear_function_data.txt"

linear_data = pd.read_csv(linear_path, header=None, delimiter=" ").values
x = linear_data[:, 0]
y = linear_data[:, 1]
A = np.vstack([x, np.ones(len(x))]).T

Approximator = LinearApprox()
# Approximator.fit(x, y)
# predict = Approximator.predict(x)
predict = Approximator.linear_approx(A, y)

a, b = Approximator.get_coe()
print('coefficients:', a)
print('bias:', b)

fig = plt.figure()
plt.plot(x, y, 'o', label='Original Data')
plt.plot(x, predict, label="Linear Approximation")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Approximate with a Linear Function')
plt.legend()
plt.show()
```

## Second Part

```python
# Plot linear approximation against original data (nonlinear)
nonlinear_path = "./data/nonlinear_function_data.txt"

nonlinear_data = pd.read_csv(nonlinear_path, header=None, delimiter=" ").values
x = nonlinear_data[:, 0]
y = nonlinear_data[:, 1]
A = np.vstack([x, np.ones(len(x))]).T

Approximator = LinearApprox()
# Approximator.fit(x, y)
# predict = Approximator.predict(x)
predict = Approximator.linear_approx(A, y)

a, b = Approximator.get_coe()
print('coefficients:', a)
print('bias:', b)

fig = plt.figure()
plt.plot(x, y, 'o', label='Original Data')
plt.plot(x, predict, '.', label="Linear Approximation")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Approximate with a Nonlinear Function')
plt.legend()
plt.show()
```

## Third Part

```python
# Plot radial basis approximation against original data (nonlinear)
nonlinear_path = "./data/nonlinear_function_data.txt"
nonlinear_data = pd.read_csv(nonlinear_path, header=None, delimiter=" ").values

x = nonlinear_data[:, 0]
y = np.expand_dims(nonlinear_data[:, 1], axis=1)
A = np.vstack([x, np.ones(len(x))]).T

Approximator = NonlinearApprox(15, 0.8)
# Approximator.fit(x, y)
# predict = Approximator.predict(x)
predict = Approximator.radial_approx(A, y)

# a, b = Approximator.get_coe()
# print('coefficients:', a)
# print('bias:', b)

fig = plt.figure()
plt.plot(nonlinear_data[:, 0], nonlinear_data[:, 1], 'o', label='Original Data')
plt.plot(nonlinear_data[:, 0], predict, '.', label="Radial Functions Approximation")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Approximate with a Combination of Radial Functions')
plt.legend()
plt.show()
```
